```python
!pip -q install --upgrade "numpy==2.0.2" "scipy==1.11.4" "gensim==4.3.3"

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os, pandas as pd, numpy as np
PROJ = "/content/drive/MyDrive/dissertation"
DATA_DIR = f"{PROJ}/data"
OUT_DIR  = f"{PROJ}/outputs/embeddings"
os.makedirs(OUT_DIR, exist_ok=True)

def make_sentences(frame: pd.DataFrame, cols):
    sents = []
    sub = frame[cols].astype(str)
    for _, row in sub.iterrows():
        sents.append([f"{c}:{row[c]}" for c in cols])  # keep prefixes to avoid collisions
    return sents

def save_vectors(keys, matrix, path_csv):
    out = pd.DataFrame(matrix, index=keys); out.index.name = "token"
    out.to_csv(path_csv)
    print("Saved:", path_csv, out.shape)
```

```python
from gensim.models import Word2Vec
from gensim.models.fasttext import FastText

DATASET = "breast"
DATA_PATH = f"{DATA_DIR}/Breast_clean.csv"
CATS = ["TNM_PATH_T","TNM_PATH_N","hospid"]

df = pd.read_csv(DATA_PATH, low_memory=False)
for c in CATS:
    df[c] = df[c].astype(str).str.strip().replace({"": np.nan})
df[CATS] = df[CATS].fillna("Unknown")

print("Breast uniques:", {c: df[c].nunique() for c in CATS}, "| rows:", len(df))
sentences = make_sentences(df, CATS)

# Word2Vec
w2v = Word2Vec(sentences=sentences, vector_size=64, window=5, min_count=1,
               workers=2, sg=1, negative=5, epochs=10)
w_keys = w2v.wv.index_to_key
w_mat  = np.vstack([w2v.wv[k] for k in w_keys])
w2v_path = f"{OUT_DIR}/{DATASET}_w2v_64d_prefixed.csv"
save_vectors(w_keys, w_mat, w2v_path)

# FastText
ft = FastText(sentences=sentences, vector_size=64, window=5, min_count=1,
              workers=2, sg=1, negative=5, epochs=10)
f_keys = ft.wv.index_to_key
f_mat  = np.vstack([ft.wv[k] for k in f_keys])
ft_path = f"{OUT_DIR}/{DATASET}_fasttext_64d_prefixed.csv"
save_vectors(f_keys, f_mat, ft_path)
```

```python
# Install Node2Vec + compatible NetworkX (won’t touch NumPy/gensim already used)
!pip -q uninstall -y node2vec networkx > /dev/null
!pip -q install "networkx==2.8.8" "node2vec==0.4.3"

import networkx as nx
from node2vec import Node2Vec

def build_graph_from_cats(frame: pd.DataFrame, cols):
    G = nx.Graph()
    rows = frame[cols].astype(str).values.tolist()
    for r in rows:
        toks = [f"{c}:{v}" for c, v in zip(cols, r)]
        for i in range(len(toks)):
            for j in range(i+1, len(toks)):
                u, v = toks[i], toks[j]
                if G.has_edge(u, v):
                    G[u][v]["weight"] += 1.0
                else:
                    G.add_edge(u, v, weight=1.0)
    return G
```

```python
DATASET = "breast"
DATA_PATH = f"{DATA_DIR}/Breast_clean.csv"
CATS = ["TNM_PATH_T","TNM_PATH_N","hospid"]

df = pd.read_csv(DATA_PATH, low_memory=False)
for c in CATS:
    df[c] = df[c].astype(str).str.strip().replace({"": np.nan})
df[CATS] = df[CATS].fillna("Unknown")

G = build_graph_from_cats(df, CATS)
print("Breast graph:", G.number_of_nodes(), "nodes,", G.number_of_edges(), "edges")

n2v = Node2Vec(G, dimensions=64, walk_length=10, num_walks=50,
               workers=2, p=1, q=1, weight_key="weight")
n2v_model = n2v.fit(window=5, min_count=1, sg=1, epochs=5)

keys = list(G.nodes())
mat  = np.vstack([n2v_model.wv[k] for k in keys])
n2v_path = f"{OUT_DIR}/{DATASET}_node2vec_64d_prefixed.csv"
save_vectors(keys, mat, n2v_path)
```

```python
# =========================
# Sanity check: all embeds
# =========================
import os, pandas as pd, numpy as np

PROJ = "/content/drive/MyDrive/dissertation"
OUT_DIR = f"{PROJ}/outputs/embeddings"

CONFIG = {
    "breast": {
        "cats": ["TNM_PATH_T","TNM_PATH_N","hospid"],
        "files": [
            "breast_w2v_64d_prefixed.csv",
            "breast_fasttext_64d_prefixed.csv",
            "breast_node2vec_64d_prefixed.csv",
        ],
        "sample_token": "TNM_PATH_T:p1A"  # adjust if missing
    },
    "adult": {
        "cats": ["occupation","workclass","native_country"],
        "files": [
            "adult_w2v_64d_prefixed.csv",
            "adult_fasttext_64d_prefixed.csv",
            "adult_node2vec_64d_prefixed.csv",
        ],
        "sample_token": "occupation:Prof-specialty"  # adjust if missing
    },
    "petfinder": {
        "cats": ["Breed1","Color1","MaturitySize"],
        "files": [
            "petfinder_w2v_64d_prefixed.csv",
            "petfinder_fasttext_64d_prefixed.csv",
            "petfinder_node2vec_64d_prefixed.csv",
        ],
        "sample_token": "Breed1:Mixed Breed"  # adjust if missing
    }
}

def nn(df, token, k=5):
    if token not in df.index:
        print(f"  ⚠ token not found: {token}. Try one of:", df.index[:5].tolist()); return
    X = df.values.astype(float)
    v = df.loc[token].values.astype(float)
    sims = (X @ v) / (np.linalg.norm(X, axis=1)*np.linalg.norm(v) + 1e-8)
    order = np.argsort(-sims)[:k+1]
    print(f"  NN for {token}:")
    for i in order:
        print(f"    {df.index[i]:<40} cos={sims[i]:.3f}")

def approx_vocab_expect(dataset):
    # rough check by reading the first (w2v) file’s index count
    path = os.path.join(OUT_DIR, f"{dataset}_w2v_64d_prefixed.csv")
    if not os.path.exists(path): return None
    E = pd.read_csv(path, index_col="token")
    return E.shape[0]

errors = 0
for ds, cfg in CONFIG.items():
    print(f"\n=== {ds.upper()} ===")
    expected_dim = 64
    for fname in cfg["files"]:
        path = os.path.join(OUT_DIR, fname)
        if not os.path.exists(path):
            print("  Missing file:", path); errors += 1; continue
        try:
            E = pd.read_csv(path, index_col="token")
            ok_dim = (E.shape[1] == expected_dim)
            print(f"  {fname}  | shape={E.shape} | dim_ok={ok_dim}")
            # peek a few tokens
            print("   sample tokens:", E.index[:5].tolist())
        except Exception as e:
            print("  Failed to read:", path, "|", e); errors += 1

    # optional quick NN check on one token (uses W2V file if present)
    w2v_path = os.path.join(OUT_DIR, f"{ds}_w2v_64d_prefixed.csv")
    if os.path.exists(w2v_path):
        E = pd.read_csv(w2v_path, index_col="token")
        token = cfg["sample_token"]
        nn(E, token, k=5)

# final summary
print("\n==============")
print("Sanity check:", "ALL GOOD" if errors == 0 else f"{errors} issue(s) found ⚠")
```
